```python
#check outliers
plt.figure(figsize=(30,30))
num=1
for col in num_cols:
    if num<=26:
        plt.subplot(7,4,num)
        sns.boxplot(x=data[col],hue=data['Attrition'])
        plt.xlabel(col,fontsize=20)
    num+=1    
plt.tight_layout()
```

```python
#calculate outliers
outliers=['MonthlyIncome','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion',
'YearsWithCurrManager']
for i in outliers:
    q1=data[i].quantile(0.25)
    q3=data[i].quantile(0.75)
    IQR=q3-q1
    lower=q1-1.5*IQR
    upper=q3+1.5*IQR
    per=len(data.loc[(data[i]<lower)|(data[i]>upper)])/len(data)
    print('***********\n')
    print('column',i)
    print('IQR is:',IQR)
    print('lower limit',lower)
    print('upper limit',upper)
    print('percentage of outliers',per)
    
```

```python
#handling outliers
list=['TotalWorkingYears','YearsInCurrentRole','YearsWithCurrManager']
for i in list:
    data.loc[(data[i]<lower)|(data[i]>upper),i]=data[i].median()
```

```python
#list out constant columns
data.drop(['EmployeeCount','Over18','StandardHours'],axis=1,inplace=True)
```

# Feature Selection

```python
x=data.drop('Attrition',axis=1)
y=data['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=40)
```

```python
#your data unbalanced we performing smote operation 
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_smote,y_smote=smote.fit_resample(x_train,y_train)
```

# Model Creation
* Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_smote,y_smote)
```

```python
#prediction
LR_pred=LR.predict(x_test)
LR_pred
```

# Evaluation

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,LR_pred))
```
